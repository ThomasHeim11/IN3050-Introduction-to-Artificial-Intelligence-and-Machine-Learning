### 1. Given the function $ f(x) = -x^{4} + 2x^{3} + 2x^{2} - x $ and it's derivative $f'(x) = {- 4 x^3 + 6 x^2 + 4 x - 1}$

```python
# Standard import and functions
# Run this cell first
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return -np.power(x, 4) + 2 * np.power(x, 3) + 2 * np.power(x, 2) - x

def df(x):
    return -4 * np.power(x, 3) + 6 * np.power(x, 2) + 4 * x - 1
```

#### a. Plot the function, and its gradient (derivative) from $x = -2$ to $x = 3$

```python
# Add your solution here
x = np.linspace(-2, 3, 100)
plt.plot(x, f(x), label="f(x)")
plt.plot(x, df(x), label="df(x)")
plt.legend()
plt.show()
```

#### b. Maximize using gradient ascent. You can try step size 0.1 and start somewhere in the range $[-2, 3]$. 

Gradient ascent: $x_{(k+1)} = x_{(k)} + \gamma \nabla f(x_{(k)})$

```python
# Add your solution here
x = np.linspace(-2, 3, 100)
gamma = 0.1
x_start = 0.5
x_new = x_start
for i in range(100):
    x_new = x_new + gamma * df(x_new)
    plt.plot(x_new, f(x_new), "ro")
plt.show()
```

##### How does the choice of starting point and step size affect the algorithm's performance?

##### Is there a starting point where the algorithm would not even be able to find a local maximum?

Starting Point Effects: 
- If tou start at x < -0.5, the algorithm will converge to the local maximum around x = -1.
- If you start at x > -0.5, it will converge to the global maximum around x= 2.
- Starting at x = -0.5 is unstable as its near a local minimum.

Step Size(gamma) Effect:
- To large: The algorithm might overshot and oscillate or diverge.
- To small: The algorithm will converge very slowly.
- Just right: Smooth convergence to the maximum. 


The algorithm fails to find the global maximum when:
1. Starting left of the local minimum(around x = -0.5), as it will converge to the local maximum.
2. Starting exactly at the local minimum, as the gradient is zero. 
3. Starting point at point where the function approaches negative infinity, as it might diverge. 

```python
# Plot function and try different starting points
x = np.linspace(-2, 3, 100)
plt.plot(x, f(x), label="f(x)")

# Try different starting points
start_points = [-1.5, 0.5, 2]
gamma = 0.1

for x_start in start_points:
    x_new = x_start
    trajectory = []
    for i in range(50):
        x_new = x_new + gamma * df(x_new)
        trajectory.append(x_new)
    plt.plot(trajectory, f(np.array(trajectory)), 'o-', label=f'Start: {x_start}')

plt.legend()
plt.title('Gradient Ascent with Different Starting Points')
plt.show()# Add your solution here

```
